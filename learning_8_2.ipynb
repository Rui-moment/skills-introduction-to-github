```python
#Numpy

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-360, 361, 10)
y = np.sin(np.radians(x))

plt.plot(x, y)
plt.show()
```

```python
#リサジュー曲線
#x=sin(2*θ), y=sin(3*θ)

import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0, np.pi * 2, 0.01)
x = np.sin(5 * a)
y = np.sin(9 * a)

plt.plot(x, y)
plt.show()
```

```python
#バラ曲線
#r=m*sin(n*θ)

import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0, np.pi * 6, 0.05)
r = 4 * np.sin(4 * a / 3 )
x = r * np.cos(a)
y = r * np.sin(a)

plt.plot(x, y)
plt.show()
```

```python
# 螺旋
# r=m*e**(n*θ),m=2, n=0.1

import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0, np.pi * 16, 0.05)
r = 2 * np.exp(0.1 * a )
x = r * np.cos(a)
y = r * np.sin(a)

plt.plot(x, y)
plt.show()
```
